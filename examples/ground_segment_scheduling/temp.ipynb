```python
from utils import sedaroLogin

sedaro = sedaroLogin()
```

```python
template = sedaro.agent_template('PP7jzzBWvzybTTYxN673TR')
```

```python
dh_subsystem = "NSgheitRQ51bMoH_95sZk"
for antenna in template.Antenna.get_all():
    name = antenna.name
    modem = template.Modem.create(name=f'{name} modem', subsystem=dh_subsystem)
```

```python
# Internet to data center
data_type = template.DataType.create(name='Mission Data')
rate = 10e9
data_center_id = 'PP8GNnLr2VFmch6dVdJrR4'
for modem in template.Modem.get_all():
    name = modem.name.replace(' modem','')
    # Create interface
    template.InternalDataInterface.create(
        name=f'{name} Internet',
        onBitRate=rate,
        alwaysActive=True,
        dataTypes={data_type.id: {'priority': 1}},
        source=modem.id,
        sink=data_center_id
    )
```

```python
to_delete = [template.DataType.get_first().id]
for interface in template.InternalDataInterface.get_all():
    to_delete.append(interface.id)

template.crud(delete=to_delete)
```

```python
abstract_modem_id = 'PP8HN7WBmFXLvxsk2zYD9r'
for interface in template.InternalDataInterface.get_all():
    antenna_name = interface.name.split(' ')[0]
    antenna_id = template.Antenna.get_where(name=antenna_name)[0].id
    
    template.InternalDataInterface.create(
        name=f'Localize {antenna_name} Data',
        onBitRate=rate,
        alwaysActive=True,
        dataTypes={data_type.id: {'priority': 1}},
        source=abstract_modem_id,
        sink=antenna_id
    )
```
